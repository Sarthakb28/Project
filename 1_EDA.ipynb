```python
import ast
import pandas as pd
import seaborn as sns
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
#Filter for US Data Analyst roles
df_DA_US=df[(df['job_title_short']=='Data Analyst') & (df['job_country']=='United States')]
df_DA_US
```

```python
df_plot=df_DA_US['job_location'].value_counts().head(10).to_frame()
df_plot
sns.barplot(data=df_plot, x='count',y='job_location', hue='count',palette='dark:b_r',legend=False)
plt.xlabel('Number of Jobs', fontsize=12)
plt.ylabel('Job Location in US', fontsize=12)
plt.title('Counts of Job Locations for Data Analyst in US', fontsize=14)

# Removing the border
sns.despine()
plt.show()
```

```python
dict_column = {
    'job_work_from_home': 'Work from Home Offered',
    'job_no_degree_mention': 'Degree Requirement',
    'job_health_insurance': 'Health Insurance Offered'
}

fig, ax = plt.subplots(1, 3, figsize=(11, 3.5))

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df_DA_US[column].value_counts(), labels=['False', 'True'], autopct='%1.1f%%', startangle=90)
    ax[i].set_title(title)

plt.show()
```

```python
df_company=df_DA_US['company_name'].value_counts().sort_values(ascending=False).head(10).to_frame()
df_company
sns.barplot(data=df_company,x='count',y='company_name',hue='count', palette='dark:b_r',legend=False)
plt.title('Counts of companies for DA in US')
plt.xlabel('Number of job postings')
plt.ylabel('Company Names')
sns.despine()
plt.show()
```
